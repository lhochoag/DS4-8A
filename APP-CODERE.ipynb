```python
import pandas as pd
from datetime import datetime

#pd_fecha = pd.re]ad_html('https://apuestas.wplay.co/es/t/19311/Colombia-Primera-A')[0]
#pd_fecha

def equipo(lista):
    
    equipo=""
    for j in range(1,len(lista)-3):
        equipo+=lista[j]+" "
    return equipo[0:len(equipo)-1]

def pronostico(pd_fecha):
    FecList=[]
    GanaList=[]
    PronList=[]
    LocalList=[]
    VisitList=[]
    CuotaLocList=[]
    CuotaEmpList=[]
    CuotaVisList=[]

    #print (pd_fecha)
    #print (pd_fecha[3])
    
    for partido in range(len(pd_fecha)):
        
    
        local_list=pd_fecha[4][partido].split() #3
        #print(local_list)
        empate_list=pd_fecha[5][partido].split() #4
        #print (empate_list)
        visitante_list=pd_fecha[6][partido].split() #5
        #print (visitante_list)
        cuota_l=local_list[-2]
        cuota_e=empate_list[-2]
        cuota_v=visitante_list[-2]
        equipo_local=equipo(local_list)
        equipo_visitante=equipo(visitante_list)
    
        loc=float(cuota_l)
        emp=float(cuota_e)
        vis=float(cuota_v)
    
        pronostico=-0.805*loc-0.0517*emp+0.0125*vis+2.3923
    
        umbral=0.33
    
        if pronostico>umbral:
            gana="L"
        elif pronostico<0:
            gana="V"
        else:
            gana="E"
    
        print (pd_fecha[1][partido],gana,round(pronostico,2),equipo_local,"-",equipo_visitante,cuota_l,cuota_e,cuota_v,"     ")
        FecList.append( pd_fecha[1][partido])
        GanaList.append(gana)
        PronList.append(pronostico)
        LocalList.append(equipo_local)
        VisitList.append(equipo_visitante)
        CuotaLocList.append(loc)
        CuotaEmpList.append(emp)
        CuotaVisList.append(vis)
        
        
    resultados= {'hoy':datetime.now().strftime("%d/%m/%Y %H:%M"),
                 'Fecha':FecList,'LEV':GanaList,
        'Dif Pron':PronList,
        'Local':LocalList,
        'Visitante':VisitList,
        'C.Local':CuotaLocList,
        'C.Empate':CuotaEmpList,
        'C.Visitante':CuotaVisList
        }                       
    df=pd.DataFrame(resultados)
    
    return df

def fileout():

    from datetime import datetime
    now = datetime.now() # current date and time
    print (now)
    year = now.strftime("%Y")
    print("year:", year)

    month = now.strftime("%m")
    print("month:", month)

    day = now.strftime("%d")
    print("day:", day)

    time = now.strftime("%H:%M:%S")
    print("time:", time)

    #date_time = now.strftime("%m/%d/%Y, %H:%M:%S")
    date_time = now.strftime("%Y%m%d-%H%M") 
    salida=date_time+"-futcol.xls"
    print (salida)
    time_out=now.strftime("%d/%m/%Y %H:%M")
    print (time_out)
    #print("date and time:",date_time)
    return salida
    
#cars = {'Brand': ['Honda Civic','Toyota Corolla','Ford Focus','Audi A4'],'Price': [22000,25000,27000,35000]}
#df = pd.DataFrame(cars, columns = ['Brand', 'Price'])
     
```
